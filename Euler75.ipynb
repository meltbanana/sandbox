```python
from math import sqrt
import pickle

max_len = 1500000
c = max_len/(sqrt(2)+1)
a = c / sqrt(2)

print(a, c, 2*a+c)

max_a = int(max_len/2) + 1
print(max_a)
```

```python
from tqdm import tqdm

sqrt_table = {}

for i in tqdm(range(1, max_a)):
  sqrt_table[i*i] = i

print(len(sqrt_table))
print(sqrt_table[625])
```

```python
with open('euler75_dump_525000.p', 'rb') as handle:
    results = pickle.load(handle)
print(len(results))
```

```python
print([v for i, v in enumerate(results.items()) if i < 10])
print(results[(40, 30, 50)])
```

```python
#results = {}
for a in tqdm(range(525001, max_a)):
  max_b = int((max_len**2 - 2*max_len*a)/(2*max_len-2*a))
  for b in range(1, min(a, max_b)):
    c = sqrt_table.get(a*a+b*b, False)
    if c:
      results[(a, b, c)] = (a+b+c)
  if a%int(max_a/10) == 0: 
    with open(f'euler75_dump_{a}.p', 'wb') as fp:
        pickle.dump(results, fp, protocol=pickle.HIGHEST_PROTOCOL)    
print(len(results))

with open('euler75results.p', 'wb') as fp:
    pickle.dump(results, fp, protocol=pickle.HIGHEST_PROTOCOL)
```
