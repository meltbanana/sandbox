```python
#debug
primes = (2, 3, 5, 7, 11)
variations = {}

for i in range(2, 11):

  variations[i] = [[i]] if i in primes else []


  for p in primes:

    if i-p<0:
      break

    if i-p in variations and i-p>=p:
      for v in variations[i-p]:
        lst = [p] + list(v)
        variations[i].append(lst)

  # 8: [[2, 2, 2, 2], [2, 3, 3], [3, 5], [3, 2, 3]] => normalization needed!
  variations[i] = set([tuple(sorted(v)) for v in variations[i]])
print(variations)
```

```python
def get_primes(max):
#Sieve of Eratosthenes
	indx = [i for i in range(max+1)]

	for i in range(len(indx)):
		if i>1 and indx[i]>0:
			count = 2
			while count*i <= max:
				indx[count*i] = 0
				count += 1

		#progress bar
		#if i%(max//10) == 0:
		#	print(i)
	primes = []
	for i in indx:
		if i>1: primes.append(i)

	return primes
```

```python
#prod

primes = get_primes(100000)
variations = {}
i = 2
results = False

while not results:

  variations[i] = [[i]] if i in primes else []

  for p in primes:

    if i-p<0:
      break

    if i-p in variations and i-p>=p:
      for v in variations[i-p]:
        lst = [p] + list(v)
        variations[i].append(lst)

  # 8: [[2, 2, 2, 2], [2, 3, 3], [3, 5], [3, 2, 3]] => normalization needed!
  variations[i] = set([tuple(sorted(v)) for v in variations[i]])

  if len(variations[i]) >= 5000:
    results = i
    break

  i += 1
print(results)
```
