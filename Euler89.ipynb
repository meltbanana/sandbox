```python
def latin_decoder(s):
    numbers = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    result = 0
    stage = 0
    lst = [numbers[x] for x in list(s)]
    for i in range(len(lst)):
        stage += lst[i]
        if i < len(lst) - 1:
            if lst[i] > lst[i+1]:
                result += stage                 
                stage = 0
            elif lst[i] < lst[i+1]:
                result -= stage                 
                stage = 0               
        else:
            result += stage           
    return result
```

```python
latin_decoder("XIX")
```

```python
def latin_encoder(n):
    result = ""
    rest = n
    
    n1000 = rest // 1000
    rest -= n1000 * 1000
    for i in range(n1000):
        result += "M"
    
    n100 = rest // 100
    rest -= n100 * 100
    dict_n100 = {0: "", 1: "C", 2: "CC", 3: "CCC", 4: "CD", 5: "D", 6: "DC", 7: "DCC", 8: "DCCC", 9: "CM"}
    
    result += dict_n100[n100]
    
    n10 = rest // 10
    rest -= n10 * 10
    dict_n10 = {0: "", 1: "X", 2: "XX", 3: "XXX", 4: "XL", 5: "L", 6: "LX", 7: "LXX", 8: "LXXX", 9: "XC"}
    result += dict_n10[n10]
    
    
    dict_n1 = {0: "", 1: "I", 2: "II", 3: "III", 4: "IV", 5: "V", 6: "VI", 7: "VII", 8: "VIII", 9: "IX"}
    result += dict_n1[rest]
 
    return result
```

```python
latin_encoder(13278)
```

```python
latin_encoder(latin_decoder("XXXXVIIII"))
```

```python
result = 0
with open('0089_roman.txt', 'r') as file:
    for line in file:
        src = line.strip()
        dest = latin_encoder(latin_decoder(src))
        result += len(src)-len(dest)
print(result)
```
