```python
from tqdm import tqdm
```

```python
def rectangles(a, b):
    result = 0
    for i in range(1, a+1):
        for k in range(1, b+1):
            result += (a-i+1) * (b-k+1)
            #print(i,k, (a-i+1), (b-k+1), (a-i+1) * (b-k+1))
    return result
```

```python
rectangles(1,2000)
```

```python
best, best_x, best_y = 0, None, None
goal = 2000000

for x in tqdm(range(5, 2000)):
    y0, y1 = 1, x
    y = (y1 + y0)//2
    while y0<=y<=y1:
        result = rectangles(x, y)
        
        if (goal-result)**2 < (goal-best)**2:
            best, best_x, best_y = result, x, y
        
        if y == y0 or y == y1:
            break
        
        if result<=goal:
            y0, y1 = y0, y
        else:
            y0, y1 = y, y1
        y = (y1 + y0)//2
        


print(best_x, best_y, best)
```
