```python
from tqdm import tqdm

def get_all_fractions(n, min_factor=2):
    result = []
    
    for d in range(min_factor, int(n**0.5)+1):
        if n % d == 0:
            result.append([d, n//d])
            for sub in get_all_fractions(n//d, d):
                result.append([d] + sub)
    return result
```

```python
print(get_all_fractions(12))
```

```python
max_n = 100000
result = {}
for n in tqdm(range(4, max_n+1)):
    lst = [(n-sum(fractions)+len(fractions)) for fractions in get_all_fractions(n)]
    for k in lst:     
        if k not in result and k <= 12000:
            result[k] = n
    

print(len(result.keys()))
print(sum(set(result.values())))

```
